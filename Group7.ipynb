```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/sondean/group7/main/BankChurners.csv"
df = pd.read_csv(url,encoding = 'unicode_escape')
df.head()
```

```python
counts = df.groupby(["Education_Level", "Card_Category"]).size().unstack()
counts.plot(kind="bar")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.title("Card Category Distribution by Education Level")
plt.show()
```

```python
x = df['Income_Category']
income_counts = df['Income_Category'].value_counts()
plt.plot(income_counts.index,income_counts,'--or')
plt.xlabel("Income_Category")
plt.ylabel("Count")
plt.title("Income Category ")
plt.show()
```

```python
plt.figure(figsize=(10,5))
dfscatter=df[['Total_Trans_Amt','Total_Trans_Ct']]
plt.scatter(dfscatter['Total_Trans_Amt'], dfscatter['Total_Trans_Ct'], color='blue', s=50, marker='1')

plt.ylabel("Total transaction count")
plt.xlabel("Total transaction amount")
plt.show()
```

```python
customer_age=df["Customer_Age"]
plt.hist(customer_age,color="orange")
plt.xlabel("Age of customers")
plt.ylabel("Number of customers")
plt.title("Distribution of customer age")
plt.show()
```
